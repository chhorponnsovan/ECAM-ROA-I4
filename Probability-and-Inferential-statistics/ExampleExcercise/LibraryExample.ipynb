```python
import importlib
import statisticlib
importlib.reload(statisticlib)
from statisticlib import regression_scatterplot, regression_summary, regression_equation, regression_intervals, correlation, anova_one_way, t_test, f_test
import pandas as pd
```

# Examples

```python
x = [2, 3, 3, 3, 4, 4, 5, 5, 5, 6]
y = [28.7, 24.8, 26.0, 30.5, 23.8, 24.6, 23.8, 20.4, 21.6, 22.1]
df = pd.DataFrame({'x': x, 'y': y})
```

```python
t_test(df,'x','y')
```

```python
f_test(df,'x','y')
```

```python
anova_one_way(df,'x','y')
```

```python
correlation(df,'y','x') # y is the dependent variable, x is the independent variable
```

```python
regression_summary(df,'y',['x'], print_summary=True)
```

```python
regression_equation(df,'y',['x'])
```

```python
regression_intervals(df,'y',['x'],predict_values=[2], alpha=0.05)
```

```python
regression_scatterplot(df, 'y', ['x'])
```
